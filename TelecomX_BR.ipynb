```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
```

#📔Dicionário de dados


*customerID*: número de identificação único de cada cliente

*Churn*: se o cliente deixou ou não a empresa

*gender*: gênero (masculino e feminino)

*SeniorCitizen*: informação sobre um cliente ter ou não idade igual ou maior que 65 anos

*Partner*: se o cliente possui ou não um parceiro ou parceira

*Dependents*: se o cliente possui ou não dependentes
tenure: meses de contrato do cliente

*PhoneService*: assinatura de serviço telefônico

*MultipleLines*: assisnatura de mais de uma linha de telefone

*InternetService*: assinatura de um provedor internet

*OnlineSecurity*: assinatura adicional de segurança online

*OnlineBackup*: assinatura adicional de backup online

*DeviceProtection*: assinatura adicional de proteção no dispositivo

*TechSupport*: assinatura adicional de suporte técnico, menos tempo de espera

*StreamingTV*: assinatura de TV a cabo

*StreamingMovies*: assinatura de streaming de filmes

*Contract*: tipo de contrato

*PaperlessBilling*: se o cliente prefere receber online a fatura

*PaymentMethod*: forma de pagamento

*Charges.Monthly*: total de todos os serviços do cliente por mês

*Charges.Total*: total gasto pelo cliente

#📌 Extracão

```python
dados = pd.read_json(url)
dados
```

#🔧 Transformação

```python
dados.head()
```

```python
# Normalize nested 'customer' column
df_customer = pd.json_normalize(dados['customer'])

# Normalize nested 'phone' column
df_phone = pd.json_normalize(dados['phone'])

# Normalize nested 'internet' column
df_internet = pd.json_normalize(dados['internet'])

# Normalize nested 'account' column
df_account = pd.json_normalize(dados['account'])
# Select the already flat columns
df_flat_main = dados[['customerID', 'Churn']]

# Concatenate all normalized parts along with the original flat columns
dados = pd.concat([df_flat_main, df_customer, df_phone, df_internet, df_account], axis=1)

dados['Churn'] = dados['Churn'].replace("", "No data")
```

#📊 Carga e análise

```python
#média de cobranças por mês, máximo e mínimo cobrado por mês

# Convert 'Charges.Monthly' and 'Charges.Total' to numeric, coercing errors to NaN
dados['Charges.Monthly'] = pd.to_numeric(dados['Charges.Monthly'], errors='coerce')
dados['Charges.Total'] = pd.to_numeric(dados['Charges.Total'], errors='coerce')

# Fill NaN values with 0. This assumes that missing charges should be treated as zero.
dados['Charges.Monthly'] = dados['Charges.Monthly'].fillna(0)
dados['Charges.Total'] = dados['Charges.Total'].fillna(0)

média_cobrança_mes = dados['Charges.Monthly'].mean()
média_cobrança_total = dados['Charges.Total'].mean()
max_cobrança_mes = dados['Charges.Monthly'].max()
min_cobrança_mes = dados['Charges.Monthly'].min()

print('O preço médio cobrado por cliente foi de ', média_cobrança_mes, ' Dólares.')
print('O preço médio cobrado por cliente no total foi de ', média_cobrança_total, ' Dólares.')
print('O máximo cobrado para um cliente foi ', max_cobrança_mes, ' Dólares.')
print('O mínimo cobrado para um cliente foi ', min_cobrança_mes, ' Dólares.')
```

```python
# Clientes que foram cobrados acima da média ao mês
clientes_acima_media_mes = dados[dados['Charges.Monthly'] > média_cobrança_mes]
clientes_acima_media_mes
```

```python
# Clientes que foram cobrados acima da média do total
clientes_acima_media_total = dados[dados['Charges.Total'] > média_cobrança_total]
clientes_acima_media_total
```

```python
# Valor mediano de pagamentos por gênero ao mês
valor_mediano_mes = dados.groupby("gender")["Charges.Monthly"].median()
valor_mediano_mes
```

```python
# Valor mediano de pagamentos por gênero no total
valor_mediano_total = dados.groupby("gender")["Charges.Total"].median()
valor_mediano_total
```

```python
desvio_cobrado_mes = dados["Charges.Monthly"].std()
desvio_cobrado_total = dados["Charges.Total"].std()

print('O desvio de Valores cobrados ao mês foi de: ', desvio_cobrado_mes)
print('O desvio de Valores cobrados no total foi de: ', desvio_cobrado_total)
```

```python
desvio_cobrado_mes_gênero = dados.groupby("gender")["Charges.Monthly"].std()
desvio_cobrado_total_gênero = dados.groupby("gender")["Charges.Total"].std()

print('O desvio de Valores cobrados ao mês, por gênero, foi de: ', desvio_cobrado_mes_gênero)
print('O desvio de Valores cobrados no total, por gênero, foi de: ', desvio_cobrado_total_gênero)
```

# Estudo de evasão

```python
taxa_churn = (dados["Churn"].value_counts(normalize=True).fillna(0) * 100).round(2)

labels = ['No', 'Yes', 'No data']

plt.figure(figsize=(9,11))

plt.pie(taxa_churn, labels=labels, autopct='%1.2f%%', startangle=90)
plt.title("Taxa de Churn")
plt.axis('equal')


plt.legend(labels, title="Churn", loc="center left")

plt.show()
```

```python
contrato_churn = estudo_churn.groupby(["Contract", "Churn"]).size().unstack()

ax = contrato_churn.plot(kind="bar", stacked=True, figsize=(8, 6))

plt.title("Churn por Tipo de Contrato")
plt.xlabel("Tipo de Contrato")
plt.ylabel("Número de Clientes")
plt.xticks(rotation=0)
ax.legend(title="Churn") # Adiciona a legenda usando o objeto Axes (ax)

plt.show()
```
